# AI Labour Market & Occupational Demand Analysis

**Executive Summary:**
- The dataset reveals that while high AI adoption environments command a significant salary premium, they are not universally concentrated in the largest enterprises; mid-sized companies offer highly competitive compensation.
- The Finance and Entertainment industries currently present the highest average salaries, but potential job seekers must weigh this against significantly elevated automation risks inherent to those sectors.
- Career future-proofing requires prioritizing cross-functional, hard-to-automate skills—like strategic project management and domain expertise—which consistently decouple high compensation from automation vulnerability.


```
import pandas as pd
import plotly.express as px
import plotly.io as pio

# Set global highly-professional dark theme
pio.templates.default = "plotly_dark"

# Ingest cleaned, structured data
df = pd.read_csv("ai_jobs_cleaned.csv")
```

**Q1 — Which industries pay the most and is it worth the automation risk?**


```
# Aggregate average salary and dominant risk level per industry
q1_df = df.groupby('Industry').agg(
    Avg_Salary=('Salary_USD', 'mean'),
    Dominant_Risk=('Automation_Risk', lambda x: x.mode()[0])
).reset_index().sort_values('Avg_Salary', ascending=True)

fig1 = px.bar(
    q1_df, 
    x='Avg_Salary', 
    y='Industry', 
    orientation='h',
    color='Dominant_Risk',
    title='Average Salary vs. Dominant Automation Risk by Industry',
    labels={'Avg_Salary': 'Average Salary (USD)', 'Dominant_Risk': 'Primary Risk Level'},
    text_auto='.3s',
    color_discrete_map={'Low': '#2ca02c', 'Medium': '#ff7f0e', 'High': '#d62728'}
)
fig1.show()
```

**Insight:** 
Finance and Entertainment command the highest baseline compensation, but both are flooded with 'High' automation risks. Education and Retail offer lower compensation but present substantially safer long-term employment propositions due to broad 'Low' to 'Medium' automation exposure.


**Q2 — Does AI adoption actually translate to higher salaries?**


```
# Ensure logical progression sorting
adoption_order = ['Low', 'Medium', 'High']
q2_df = df.groupby('AI_Adoption_Level')['Salary_USD'].mean().reset_index()
q2_df['AI_Adoption_Level'] = pd.Categorical(q2_df['AI_Adoption_Level'], categories=adoption_order, ordered=True)
q2_df = q2_df.sort_values('AI_Adoption_Level')

fig2 = px.bar(
    q2_df, 
    x='AI_Adoption_Level', 
    y='Salary_USD', 
    title='Average Salary Premium by Corporate AI Adoption Maturity',
    labels={'AI_Adoption_Level': 'Corporate AI Adoption Level', 'Salary_USD': 'Average Salary (USD)'},
    text_auto='.3s',
    color='AI_Adoption_Level',
    color_discrete_sequence=['#4E79A7', '#F28E2B', '#E15759']
)
fig2.show()
```

**Insight:** 
There is a clear, institutional salary premium associated with high AI adoption maturity. Companies that aggressively integrate AI into their operations consistently compensate their labor force at higher market rates than those lagging in adoption.


**Q3 — What skills should someone prioritize to future-proof their career?**


```
# Identify highest frequency skills and map them against overarching automation risk
q3_df = df['Required_Skills'].value_counts().head(10).reset_index()
q3_df.columns = ['Skill', 'Frequency']

risk_by_skill = df.groupby('Required_Skills')['Automation_Risk'].agg(lambda x: x.mode()[0]).reset_index()
q3_df = q3_df.merge(risk_by_skill, left_on='Skill', right_on='Required_Skills', how='left')

fig3 = px.bar(
    q3_df.sort_values('Frequency', ascending=True), 
    x='Frequency', 
    y='Skill', 
    orientation='h',
    color='Automation_Risk',
    title='Top 10 High-Demand Skills vs Vulnerability Risk',
    labels={'Frequency': 'Market Frequency', 'Skill': 'Required Skill'},
    text='Frequency',
    color_discrete_map={'Low': '#2ca02c', 'Medium': '#ff7f0e', 'High': '#d62728'}
)
fig3.show()
```

**Insight:** 
While Python and basic Data Analysis dominate the sheer volume of demand, they frequently heavily overlap with roles subject to high automation risk. To truly build a resilient career, practitioners must prioritize hard-to-automate strategic frameworks—like overarching project execution and domain specialization.


**Q4 — Do larger companies actually pay more or is that a myth?**


```
size_order = ['Small', 'Medium', 'Large']
q4_df = df.groupby('Company_Size')['Salary_USD'].mean().reset_index()
q4_df['Company_Size'] = pd.Categorical(q4_df['Company_Size'], categories=size_order, ordered=True)
q4_df = q4_df.sort_values('Company_Size')

fig4 = px.bar(
    q4_df, 
    x='Company_Size', 
    y='Salary_USD', 
    title='Validation of Wage Disparities by Enterprise Size',
    labels={'Company_Size': 'Enterprise Size', 'Salary_USD': 'Average Pipeline Salary (USD)'},
    text_auto='.3s',
    color='Company_Size',
    color_discrete_sequence=['#59A14F', '#76B7B2', '#4E79A7']
)
fig4.show()
```

**Insight:** 
The widespread narrative that massive enterprise corporations strictly dictate the top of the wage market is a myth. The data indicates that mid-sized and smaller agile outfits aggressively compete for specialized AI talent, heavily closing the compensation gap.


**Q5 — Which locations offer the best salary opportunities?**


```
q5_df = df.groupby('Location')['Salary_USD'].mean().reset_index().sort_values('Salary_USD', ascending=True).tail(10)

fig5 = px.bar(
    q5_df, 
    x='Salary_USD', 
    y='Location', 
    orientation='h',
    title='Top 10 Global Target Locations by Average Compensation',
    labels={'Salary_USD': 'Average Salary (USD)', 'Location': 'Global Hub'},
    text_auto='.3s',
    color='Salary_USD',
    color_continuous_scale='blues'
)
fig5.show()
```

**Insight:** 
A geographically diversified approach is optimal for high earners. While established traditional tech hubs continue to provide robust baseline premiums, emerging global and hybridized centers present lucrative targets without the extreme local density constraints.


**The "So What" — Actionable Intelligence for Job Seekers:**
- **Target High-Adoption Operations:** Exclusively prioritize firms with mature AI adoption frameworks to secure maximum baseline compensation.
- **Hedge Against Vertical Risk:** If pursuing high-paying but high-risk industries (like Finance), deliberately index your skill set toward low-risk strategic capabilities rather than purely technical execution.
- **Size Doesn't Dictate Salary:** Discard the bias toward massive enterprise firms; heavily vet 'Small' and 'Medium' outfits, as they exhibit highly competitive agility in market rates.
- **Diversify the Geographic Search:** Broaden your target locations, as macro shifts have significantly flattened historical geographic compensation disparities.
- **Evolve Beyond Tooling:** Treat functional programming languages as table stakes. Differentiate your market value by pairing strict technical acumen with domain architecture and high-level project strategy.
